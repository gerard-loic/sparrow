## Notations
- B[1-9] : Bambous
- C[1-9] : Cracks
- D[1-9] : Dots
- G      : Green dragon
- R      : Red dragon
- 0      : White dragon
- N      : North wind
- E      : Est wind
- S      : South wind
- W      : West wind
- F      : Flower
- J      : Joker

## conventions d'écriture
- a()    : ensemble couleur 1
- b()    : ensemble couleur 2
- c()    : ensemble couleur 3
- d      : n'importe quel dragon
- <2468> : n'importe quel de ces chiffres, mais tous les groupes dans la mains doivent avoir le même nombre
- <1234567|1> : n'importe quel de ces chiffres, avec +1 (fonctionne obligatoirement avec <1234567>)
